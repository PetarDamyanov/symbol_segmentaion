```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("img/conn_word.jpg",cv2.IMREAD_GRAYSCALE)
plt.imshow(img, "gray")
```

```python
text = img[300:600:,0::]
formula = img[600:700:,350:550:]
plt.subplot(121)
plt.imshow(text,"gray")
plt.subplot(122)
plt.imshow(formula,"gray")
```

```python
bin_text = np.copy(text)
bin_formula = np.copy(formula)

# print(bin_formula)
treshold_t = 128
treshold_f = 32

bin_text[bin_text>treshold_t] = 255
bin_text[bin_text<=treshold_t] = 0
bin_formula[bin_formula>treshold_f] = 255
bin_formula[bin_formula<=treshold_f] = 0

plt.subplot(121)
plt.imshow(bin_text,"gray")
plt.subplot(122)
plt.imshow(bin_formula,"gray")
```

```python
conn_word = bin_text[60:120:,50:500:]
plt.imshow(conn_word, "gray")
```

```python
vertical_hist = np.sum(conn_word,axis=0,keepdims=True)/255
print(np.shape(vertical_hist))
v= list()
for i in range(np.shape(vertical_hist)[1]):
    v.append(vertical_hist[0][i])
    
plt.subplot(211)
plt.plot(v)
plt.subplot(212)
plt.imshow(conn_word, "gray")
```

```python
word_segments = list()
start = False
i1, i2 = 0, 0
m = max(v)
print(m)
for i in range(np.shape(conn_word)[1]):
    if vertical_hist[0][i] < m:
            if start == False:
                i1, start = i , True
    else:
        if start:
            i2, start = i,  False
            word_segments.append([i1,i2])

clean=list()
for word in word_segments:
    if word[1]-word[0]<=10:
        clean.append(word)
for word in clean:
    word_segments.remove(word)
print(word_segments)
```

```python
for word in word_segments:
    print(word[1]-word[0])
```

```python
min_char = word_segments[0][1]-word_segments[0][0]
for word in word_segments:
    if min_char > word[1]-word[0]:
        min_char = word[1]-word[0]
print(min_char)
```

```python
new_segment = list()
remove_segment=list()
# print(word_segments)
for word in word_segments:
#     print(word)
    if word[1]-word[0] > min_char*2:
        n = (word[1]-word[0])//min_char
        start = word[0]
        remove_segment.append(word)
        for i in range(n):
            new_segment.append([start, start+min_char])
            start+=min_char+1

# print(new_segment)
for segment in remove_segment:
    word_segments.remove(segment)
for segment in new_segment:
    word_segments.append(segment)

word_segments.sort()    
print(word_segments)
```

```python
numbers = list()

#using python slicing
for c in word_segments:
    numbers.append(conn_word[::,c[0]:c[1]:])
    
fig, axes = plt.subplots(nrows = len(numbers))
for i in range(len(numbers)):
    axes[i].imshow(numbers[i], "gray")
```
