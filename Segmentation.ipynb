```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("img/letter-to-santa-claus.png", cv2.IMREAD_GRAYSCALE)

#binary image
bin = np.copy(img)
treshold = 32
bin[bin<=treshold] = 0
bin[bin>treshold] = 255

#make horizontal hist
horizontal_hist = np.sum(bin,axis=1,keepdims=True)/255

fig, (ax1,ax2) = plt.subplots(1,2)

ax1.imshow(bin,"gray")
ax2.plot(horizontal_hist,range(np.shape(horizontal_hist)[0]))
ax2.invert_yaxis()
plt.show()
```

```python
# a more visual example of horizonatal hist
img = cv2.imread("img/letter-to-santa-claus.png", cv2.IMREAD_GRAYSCALE)

#binary image
bin = np.copy(img)
treshold = 32
bin[bin<=treshold] = 0
bin[bin>treshold] = 255

#make horizontal hist
horizontal_hist = np.sum(bin,axis=1,keepdims=True)/255

fig, ax1 = plt.subplots()

ax1.imshow(bin,"gray")
ax1.plot(horizontal_hist,range(np.shape(horizontal_hist)[0]))
plt.show()
# as we can see the hisher value we have the lesser positive pixeal we have
```

```python
#we can use max to find the most background pixels
# since max is the best value it will work well without noise 
#since we usually have some noise we need to expand the treshold according 
m = 0
for i in range(np.shape(bin)[0]):
    if horizontal_hist[i][0]>m:
        m = horizontal_hist[i][0]
print(m)
```

```python
line_segments = list()
start = False
i1, i2 = 0, 0
for i in range(np.shape(bin)[0]):
    if horizontal_hist[i][0] < m-15:
            if start == False:
                i1, start = i , True
    else:
        if start:
            i2, start = i,  False
            line_segments.append([i1,i2])


#clean up
for line in line_segments:
    if line[1]-line[0]<10:
        line_segments.remove(line)
        
print(line_segments)
```

```python
cut_parts = list()
r, c = np.shape(line_segments)
for i in range(r):
    cut_parts.append(bin[line_segments[i][0]:line_segments[i][1]:])
```

```python
fig, axes = plt.subplots(nrows = r)
for i in range(r):
    axes[i].imshow(cut_parts[i], "gray")
plt.savefig("lines.png")
```

```python
#we will continue with splitting only the first line 
line = cut_parts[0]
plt.imshow(line, "gray")
```

```python
vertical_hist = np.sum(line,axis=0,keepdims=True)/255
v = list()
for i in range(np.shape(line)[1]):
    v.append(vertical_hist[0][i])

plt.subplot(211)
plt.plot(v)
plt.subplot(212)
plt.imshow(line, "gray")
# plt.savefig("split_word_sym.png")
```

```python
word_segments = list()
start = False
i1, i2 = 0, 0
for i in range(np.shape(line)[1]):
    if vertical_hist[0][i] > 60:
            if start == False:
                i1, start = i , True
            elif start and i == np.shape(line)[1]-1:
                i2, start = i,  False
                word_segments.append([i1,i2])

    else:
        if start:
            i2, start = i,  False
            word_segments.append([i1,i2])

print(word_segments)

#clear
remove_indexs=list()
for i in range(np.shape(word_segments)[0]):
    if word_segments[i][1]-word_segments[i][0]<40:
        remove_indexs.append(i)
remove_indexs = remove_indexs[::-1]

for i in remove_indexs:
    word_segments.remove(word_segments[i])
print(word_segments)

word_segments_cut = list()

for i in range(len(word_segments)-1):
    word_segments_cut.append([word_segments[i][1],word_segments[i+1][0]])

print(word_segments_cut)

cut_parts = list()
r,c = np.shape(line)
# print(r,c)

for k in range(np.shape(word_segments_cut)[0]):
    l1 = list()
    for i in range(r):
        l2=list()
        for j in range(c):
            if j >= word_segments_cut[k][0] and j<= word_segments_cut[k][1]:
                l2.append(line[i][j])
        l1.append(l2)
    cut_parts.append(l1)

fig, axes = plt.subplots(nrows = np.shape(cut_parts)[0])
for i in range(np.shape(cut_parts)[0]):
    axes[i].imshow(cut_parts[i], "gray")
```

```python
word = cut_parts[0]
plt.imshow(word,"gray")
```

```python
vertical_hist = np.sum(word,axis=0,keepdims=True)/255

v = list()
for i in range(np.shape(word)[1]):
    v.append(vertical_hist[0][i])

plt.subplot(211)
plt.plot(v)
plt.subplot(212)
plt.imshow(word, "gray")
plt.savefig("split_word_sym.png")
```

```python
print(max(v))

char_segments = list()
start = False
i1, i2 = 0, 0
for i in range(np.shape(word)[1]):
    if vertical_hist[0][i] < max(v):
            if start == False:
                i1, start = i , True
    else:
        if start:
            i2, start = i,  False
            char_segments.append([i1,i2])

print(char_segments)

cut_parts = list()
r,c = np.shape(word)

for k in range(np.shape(char_segments)[0]):
    l1 = list()
    for i in range(r):
        l2=list()
        for j in range(c):
            if j >= char_segments[k][0] and j<= char_segments[k][1]:
                l2.append(word[i][j])
        l1.append(l2)
    cut_parts.append(l1)

fig, axes = plt.subplots(nrows = len(cut_parts))
for i in range(len(cut_parts)):
    axes[i].imshow(cut_parts[i], "gray")
```
